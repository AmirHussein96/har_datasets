```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as pl 
import seaborn as sns 
import pickle
from os.path import join, pardir

%matplotlib inline

sns.set_context('paper')
sns.set_style('darkgrid')

pwd = join(pardir, 'data', 'raw', 'pamap2')
pwd = join(pardir, 'data', 'raw', 'anguita2013')  
```

```python
def load(pwd, fname): 
    return pickle.load(open(join(pwd, f'{fname}.pkl'), 'rb'))
```

```python
labels = load(pwd, 'labels')
index = load(pwd, 'index')
folds = load(pwd, 'fold')

gyro = load(pwd, 'gyro-waist')
accel = load(pwd, 'accel-waist')
accel.loc[:, 'mag'] = (
    accel.x ** 2 + accel.y ** 2 + accel.z ** 2
).pow(0.5)

print(labels.shape, folds.shape, accel.shape, index.shape)

rr = np.arange(1000) + 10000 * 1

fig, axes = pl.subplots(2, 1, sharex=True, sharey=True, figsize=(20, 10))
accel.iloc[rr].plot(ax=axes[0])
gyro.iloc[rr].plot(ax=axes[1])
pl.tight_layout()
```
